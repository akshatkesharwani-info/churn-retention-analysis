```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

https://www.kaggle.com/datasets/maqeel732/wa-fn-usec-telco-customer-churn-csv (Data set link )

```python
df= pd.read_csv('/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
#cheaking shape
df.shape
```

```python
#review data set
df
```

```python
#cheaking Data types etc
df.info()
```

```python
#fixing data type of Total charges
df['TotalCharges']=pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
#cheaking again
df.info()
```

```python
#cheaking columns names
df.columns
```

EDA=*Exploratory Data Analysis*

```python
#cheaking NUll Value
df.isnull().sum()
```

```python
#filling Null with Median
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
#cheaking duplicate
df.duplicated().sum()
```

```python
#Converting churn values into numeric
df['Churn'] = df['Churn'].map({'Yes': 1, 'No': 0})
```

```python
#cheaking Churn mean
df['Churn'].mean()
```

```python
#churn by contract
df.groupby('Contract')['Churn'].mean()
```

```python
#churn by Tenure
df.groupby('tenure')['Churn'].mean()
```

```python
#cheaking monthly charges as per churn
df.groupby('Churn')['MonthlyCharges'].mean()
```

```python
#Overall Churn Insight
df['Churn'].value_counts(normalize=True) * 100
```

```python
#Churn by Gender
df.groupby('gender')['Churn'].mean()
```

```python
#Churn by Senior Citizen
df.groupby('SeniorCitizen')['Churn'].mean()
```

```python
#Churn by Payment Method
df.groupby('PaymentMethod')['Churn'].mean()
```

```python
#churn by internet service
df.groupby('InternetService')['Churn'].mean()
```

```python
#cheaking total charges as per churn
df.groupby('Churn')['TotalCharges'].mean()
```

```python
#visualisation of churn
sns.countplot(x='Churn', data=df)
plt.show()
```

```python
#visulation of Churn by contract
sns.countplot(x='Contract', hue='Churn', data=df)
plt.show()
```

```python
#visulation of Churn by Tenure
sns.histplot(data=df, x='tenure', hue='Churn', bins=30)
plt.show()
```

```python
#visulation of monthly charges as per churn
sns.boxplot(x='Churn', y='MonthlyCharges', data=df)
plt.show()
```

 Key Findings:
Month-to-month customers churn the most
New customers (low tenure) leave quickly
High monthly charges increase churn risk
Certain services (like tech support) reduce churn
